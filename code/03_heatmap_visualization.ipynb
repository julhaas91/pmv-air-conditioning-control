# Heat map visualization of air velocity / temperature

```python
import os,sys,inspect
current_dir = os.path.dirname(os.path.abspath(inspect.getfile(inspect.currentframe())))
parent_dir = os.path.dirname(current_dir)
sys.path.insert(0, parent_dir)

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import datetime

current_dir
```

```python
humidity_values = [30, 35, 40, 45, 50, 55, 60, 65, 70, 75, 80]

for humidity in humidity_values:
    
    filename = "2020-07-05_PMV_result_humidity" + str(humidity) + ".csv"
    import_filepath = "../../docs/_training_data/"
    export_filepath = "../../docs/_visualizations/"

    data = pd.read_csv(import_filepath + filename)#, index_col='Air_velocity/Temperature')

    # use only the values 0, 0.15, 0.30, 0.45 from the given dataframe
    
    data = data[(data['Air_velocity/Temperature']==0)
                |(data['Air_velocity/Temperature']==0.15)
                |(data['Air_velocity/Temperature']==0.30)
                |(data['Air_velocity/Temperature']==0.45)]

    data.set_index('Air_velocity/Temperature', inplace=True)

    plt.rcParams['figure.figsize'] = 15, 10
    ax = sns.heatmap(data, center=0, annot=True,cbar_kws={'label': 'PMV'})
    ax.invert_yaxis()
    #ax.set_title('PMV = f(air velocity, temperature)\nHumidity ' + str(humidity) + '%')

    plt.title('Humidity ' + str(humidity) + '%')#, fontsize = 18) # title with fontsize 20

    plt.xlabel('temperature')#, fontsize = 15) # x-axis label with fontsize 15
    plt.ylabel('air velocity')#, fontsize = 15) # y-axis label with fontsize 15

    plt.show()

    #plt.savefig(export_filepath + str(datetime.date.today()) + '_heatmap_PMV_humidity_' + str(humidity), bbox_inches='tight')
    plt.close()
```

```python
# VISUALIZE DELTA PMV

humidity_values = [30, 35, 40, 45, 50, 55, 60, 65, 70, 75, 80]

for humidity in humidity_values:
    
    filename = "2020-07-05_dPMV_result_humidity" + str(humidity) + ".csv"
    import_filepath = "../../docs/_training_data/"
    export_filepath = "../../docs/_visualizations/"

    data = pd.read_csv(import_filepath + filename)#, index_col='Air_velocity/Temperature')

    # use only the values 0, 0.15, 0.30, 0.45 from the given dataframe
    
    data = data[(data['Air_velocity/Temperature']==0)
                |(data['Air_velocity/Temperature']==0.15)
                |(data['Air_velocity/Temperature']==0.30)
                |(data['Air_velocity/Temperature']==0.45)]

    data.set_index('Air_velocity/Temperature', inplace=True)

    plt.rcParams['figure.figsize'] = 15, 10
    ax = sns.heatmap(data, center=0, annot=True,cbar_kws={'label': 'delta PMV'})
    ax.invert_yaxis()
    #ax.set_title('PMV = f(air velocity, temperature)\nHumidity ' + str(humidity) + '%')

    plt.title('Humidity ' + str(humidity) + '%')#, fontsize = 18) # title with fontsize 20

    plt.xlabel('temperature')#, fontsize = 15) # x-axis label with fontsize 15
    plt.ylabel('air velocity')#, fontsize = 15) # y-axis label with fontsize 15

    plt.show()

    #plt.savefig(export_filepath + str(datetime.date.today()) + '_heatmap_dPMV_humidity_' + str(humidity), bbox_inches='tight')
    plt.close()
```
